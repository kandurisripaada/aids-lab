```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("train.csv")

df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Cabin'].fillna(df['Cabin'].mode()[0],inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
print(df.isna().sum())
```

```python
sns.barplot(data=df,x='Pclass',y='Survived',errorbar=None,palette=['#FF7722','white','green'],edgecolor="black")
plt.title("Survival rate by passenger class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
sns.scatterplot(data=df,x='Age',y='Fare',palette=['yellow','blue'],hue='Survived')
plt.title("age vs fare")
plt.xlabel("age")
plt.ylabel("fare")
plt.show()
```

```python
dat = df[['Age','Fare','Pclass']].copy()
plt.figure(figsize=(12,6))
sns.pairplot(data=dat,hue='Pclass',palette=['Blue','red','#4A412A'])
plt.show()
```
