```python
import pandas as pd
df=pd.read_csv('Automobile.csv')
# print(df.head())
print(df.isna().sum())
```

```python
# cols=df.columns
# for col in cols:
# print(df[col].dtype)
df.head(12)
```

```python

categorical_columns = ['make', 'fuel_type', 'aspiration', 'number_of_doors', 'body_style', 'drive_wheels', 'engine_location', 'fuel_system']

for col in categorical_columns:
    df[col] = df[col].str.lower()  
    df[col] = df[col].str.replace(r'[^a-z0-9\s]', '', regex=True) 

print(df[categorical_columns].head())
```

```python
# Replace '?' with appropriate replacements in categorical columns
df['make'].replace('?', 'unknown', inplace=True)
df['fuel_type'].replace('?', 'gas', inplace=True)
df['aspiration'].replace('?', 'std', inplace=True)
df['number_of_doors'].replace('?', 'four', inplace=True)
df['body_style'].replace('?', 'sedan', inplace=True)
df['drive_wheels'].replace('?', 'fwd', inplace=True)
df['engine_location'].replace('?', 'front', inplace=True)
df['fuel_system'].replace('?', 'mpfi', inplace=True)

df['price'] = pd.to_numeric(df['price'], errors='coerce')  # Convert to numeric
df['price'].fillna(df['price'].median(), inplace=True)  # Replace NaN and ?

# Similarly, for other numerical columns
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['horsepower'].fillna(df['horsepower'].median(), inplace=True)

```

```python
df['number_of_cylinders'] = df['number_of_cylinders'].replace({
    'four': 4,
    'six': 6,
    'eight': 8
})
```

```python
# Standardize 'fuel_system' categories
df['fuel_system'] = df['fuel_system'].replace({
    'spdi': 'mpfi',
    'multi-point fuel injection': 'mpfi',
    'single-point fuel injection': 'spfi'
})

print(df[categorical_columns + ['number_of_cylinders', 'fuel_system']].head())
```

```python
# converting hoser power and peak rpm to integers
df['horsepower'] = df['horsepower'].astype(int)
df['peak_rpm'] = df['peak_rpm'].astype(int)
# Convert to numeric for price, invalid becomes NaN
df['price'] = pd.to_numeric(df['price'], errors='coerce')
# print(df.isna().sum())
df['horsepower_kW'] = df['horsepower'] * 0.7457

df['horsepower_kW']
```

```python
# Power-to-weight ratio = horsepower / curb_weight
df['power_to_weight'] = df['horsepower'] / df['curb_weight']

# Engine efficiency = horsepower / engine_size
df['engine_efficiency'] = df['horsepower'] / df['engine_size']

# Fuel efficiency = city_mpg / highway_mpg
df['fuel_efficiency'] = df['city_mpg'] / df['highway_mpg']

# Convert compression_ratio into categories
def categorize_compression_ratio(x):
    if x < 9:
        return 'Low'
    elif 9 <= x <= 11:
        return 'Medium'
    else:
        return 'High'

# Apply the function to the compression_ratio column
df['compression_ratio_category'] = df['compression_ratio'].apply(categorize_compression_ratio)

# Display the first few rows of the new columns
print(df[['power_to_weight', 'engine_efficiency', 'fuel_efficiency', 'compression_ratio_category']].head())
```
